```python
# Calculate the number of unique price points and price range for each product
price_stats = data.groupby('product_id')['selling_price'].agg(['nunique', 'min', 'max'])
price_stats['price_range'] = price_stats['max'] - price_stats['min']
price_stats.columns = ['unique_price_points', 'min_price', 'max_price', 'price_range']

# Combine metrics: Let's give equal weight to unique price points and price range
price_stats['score'] = price_stats['unique_price_points'] + price_stats['price_range']

```

```python
# Select the top products with the highest score
top_products = price_stats.sort_values(by='score', ascending=False).head(1).index.tolist()

# Separate validation data for these two products
top_product_data = data[data['product_id'].isin(top_products)]
```

# Active learning implementation

```python
sub_val
```

```python
# Define features and target variable
X_pool_AL = sub_val[feature_columns]
y_pool_AL = sub_val['log_sales']

# Initialize an empty labeled dataset
labeled_data_AL = pd.DataFrame(columns=sub_val.columns)
```

```python
# Active learning parameters
n_iterations = 64  # Number of iterations for active learning
r2_scores_AL = []
rmse_scores_AL = []
mape_scores_AL = []
```

```python
# Function to calculate evaluation metrics
def calculate_metrics(y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = mean_absolute_percentage_error(y_true, y_pred)
    return r2, rmse, mape
```

```python
# Define function to calculate prediction variance using an ensemble of Linear Regression models
def get_prediction_variance(X, y, X_pool, model, n_models=10):
    predictions = np.zeros((X_pool.shape[0], n_models))
    for i in range(n_models):
        sample_X, sample_y = resample(X, y)
        model.fit(sample_X, sample_y)
        predictions[:, i] = model.predict(X_pool)
    variance = np.var(predictions, axis=1)
    return variance
```

```python
X_combined_AL = X_train
y_combined_AL = y_train
sub_val_AL = sub_val
model_AL = model
```

```python
# Initialize a list to store predictions from each iteration
predictions_over_iterations_AL = []
```

```python
from sklearn.utils import resample
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error
# Active Learning Loop using Uncertainty Sampling
for iteration in range(n_iterations):
    # Predict the pool using the pre-trained model
    prediction_variance = get_prediction_variance(X_combined_AL.values, y_combined_AL.values, X_pool_AL.values, model_AL, 10)
    # Select the sample with the highest prediction variance
    query_idx = np.argmax(prediction_variance)
    
    #Add the instance to labeled data
    new_data_AL = sub_val_AL.iloc[query_idx].to_frame().T
    new_data_AL['log_sales'] = y_pool_AL.iloc[query_idx]
    labeled_data_AL = pd.concat([labeled_data_AL, new_data_AL])
    X_label_AL = labeled_data_AL[feature_columns]
    
    # Re-train the model
    X_combined_AL = pd.concat([X_train, X_label_AL])
    y_combined_AL = pd.concat([y_train, labeled_data_AL['log_sales']])
    model_AL.fit(X_combined_AL, y_combined_AL)
    
    # Evaluate the model on the labeled data
    if len(labeled_data_AL) > 0:
        y_pred_labeled = model_AL.predict(sub_test[feature_columns])
        r2_after, rmse_after, mape_after = calculate_metrics(sub_test['log_sales'], y_pred_labeled)
        r2_scores_AL.append(r2_after)
        rmse_scores_AL.append(rmse_after)
        mape_scores_AL.append(mape_after)
        print(f"Iteration {iteration + 1} complete. R-squared: {r2_after}, RMSE: {rmse_after}, MAPE: {mape_after}")
    else:
        r2_after = None
    
    # Predict on the entire pool and test set
    predictions_pool = model_AL.predict(sub_val[feature_columns])
    predictions_test = model_AL.predict(sub_test[feature_columns])
    
    # Store the predictions as a time series
    predictions_combined = np.concatenate([predictions_pool, predictions_test])
    predictions_over_iterations_AL.append(predictions_combined)
    
    # Remove selected sample from the pool
    sub_val_AL = sub_val_AL.drop(sub_val_AL.index[query_idx],axis = 0)
    X_pool_AL = sub_val_AL[feature_columns]
    y_pool_AL = sub_val_AL['log_sales']
```

```python
# Combine pool and test set for plotting

# Plot the actual sales
plt.figure(figsize=(14, 7))
plt.plot(sub_test.index, sub_test['sales_units'], label='Actual Sales', color='black', linewidth=2)
# Plot predictions from each iteration
for i in range(0,15):
    plt.plot(sub_test.index, np.exp(predictions_over_iterations_AL[i]), label=f'Iteration {i + 1}', linestyle='--', marker='o')

# Finalize plot
plt.title('Time Series of Predicted Sales Over Active Learning Iterations')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend(ncol = 2,loc='best')
plt.show()
```

```python
# Function to calculate the distance from the pool to the labeled outputs
def calculate_distance_to_labeled(X_pool, label, model):
    # Predict the output for  pool points
    y_pred_pool = model.predict(X_pool)
    
    # Compute the distances
    distances = []
    for pool_pred in y_pred_pool:
        # Calculate distance to all labeled predictions
        dist = np.min([np.linalg.norm(pool_pred - label_pred) for label_pred in label])
        distances.append(dist)
    return np.array(distances)
```

```python
X_combined_GS = X_train
y_combined_GS = y_train
sub_val_GS = sub_val
model_GS = model
labeled_data_GS = pd.DataFrame(columns=sub_val.columns)
```

```python
# Initialize a list to store predictions from each iteration
predictions_over_iterations_GS = []
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error
# Active Learning Loop using Uncertainty Sampling
for iteration in range(n_iterations):
    # Calculate the distance from pool to labeled outputs
    distances = calculate_distance_to_labeled(X_pool_GS, X_combined_GS.values, model_GS)
    # Select the sample with the maximum distance
    query_idx = np.argmax(distances)
    
    #Add the instance to labeled data
    new_data_GS = sub_val_GS.iloc[query_idx].to_frame().T
    new_data_GS['log_sales'] = y_pool_GS.iloc[query_idx]
    labeled_data_GS = pd.concat([labeled_data_GS, new_data_GS])
    X_label_GS = labeled_data_GS[feature_columns]
    
    # Re-train the model
    X_combined_GS = pd.concat([X_train, X_label_GS])
    y_combined_GS = pd.concat([y_train, labeled_data_GS['log_sales']])
    model_GS.fit(X_combined_GS, y_combined_GS)
    
    # Evaluate the model on the labeled data
    if len(labeled_data_GS) > 0:
        y_pred_labeled = model_GS.predict(sub_test[feature_columns])
        r2_after, rmse_after, mape_after = calculate_metrics(sub_test['log_sales'], y_pred_labeled)
        r2_scores_GS.append(r2_after)
        rmse_scores_GS.append(rmse_after)
        mape_scores_GS.append(mape_after)
        print(f"Iteration {iteration + 1} complete. R-squared: {r2_after}, RMSE: {rmse_after}, MAPE: {mape_after}")
    else:
        r2_after = None
    
    # Predict on the entire pool and test set
    predictions_test = model_GS.predict(sub_test[feature_columns])
    
    # Store the predictions as a time series
    predictions_over_iterations_GS.append(predictions_test)
    
    # Remove selected sample from the pool
    sub_val_GS = sub_val_GS.drop(sub_val_GS.index[query_idx],axis = 0)
    X_pool_GS = sub_val_GS[feature_columns]
    y_pool_GS = sub_val_GS['log_sales']
```
